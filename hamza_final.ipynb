```
import xlsxwriter
import itertools

import datetime
from collections import deque

person=["L.Sirine", "A.Marwa", "S.Rahma", "M.Tarek", "S.Ramzi", "B.Walid","K.Hamza","A.Ahmed", "S.Slim", "B.Sirine"]
poste = ["Bloc","Bloc","Bloc", "Consultation","Consultation","Consultation","Consultation", "Exploration","Exploration", "Service"]
Semaine = ["lundi","mardi","mercredi","jeudi","vendredi"]
affectedPerson = []
rotatedPoste=deque(poste)
count=0


new_list = [["Jour","Mois","Année"]]

del new_list[0]
for patient in range(0,27):
    today = datetime.date.today()
    lastMonday= today + datetime.timedelta(days=-today.weekday(), weeks=patient)
    #lastMondayClean=lastMonday.day , lastMonday.month, lastMonday.year
    #print("la semaine du" ,lastMonday.day ,"/", lastMonday.month,"/", lastMonday.year)
    new_date = [lastMonday.day,lastMonday.month, lastMonday.year]
    new_list.append(new_date)

    for jour in Semaine: 
        new_list.append([jour])
#Rotation des postes tout les jours        
        for i in person:                       
            lele=person.index(i)
            arr=[person[lele]]+[rotatedPoste[lele]]
            new_list.append(arr)        
        rotatedPoste.rotate(3)
        
#Rotation tout les 15 jours
    count=count+1
    if count % 2 ==0:        
        rotatedPoste.rotate(3)
        
#Affichage des résultats sur la console       
for row in new_list:
    print(" | ".join([str(cell) for cell in row]))
    
#Export vers un fichier Excel 
with xlsxwriter.Workbook('test.xlsx') as workbook:
    worksheet = workbook.add_worksheet()

    for row_num, data in enumerate(new_list):
       worksheet.write_row(row_num, 0, data)
```
